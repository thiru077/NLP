```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import AutoModel, AutoTokenizer

def get_attention_scores(text, model_name='bert-base-uncased'):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name, output_attentions=True)

    inputs = tokenizer(text, return_tensors='pt')
    with torch.no_grad():
        outputs = model(**inputs)

    attentions = torch.stack(outputs.attentions).squeeze(1)  # (layers, heads, seq_len, seq_len)
    avg_attention = attentions.mean(dim=[0, 1])  # Average over layers and heads
    tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'].squeeze())
    return tokens, avg_attention

def plot_attention_heatmap(tokens, attention_matrix):
    plt.figure(figsize=(10, 5))
    sns.heatmap(attention_matrix.numpy(), xticklabels=tokens, yticklabels=tokens, cmap='Blues', annot=True)
    plt.title("Attention Heatmap")
    plt.show()

# Example Medical Notes
texts = [
    "Patient reports persistent cough, high fever, and difficulty breathing for the past three days.",
    "Mild headache and occasional dizziness, but no fever or cough.",
    "Sudden loss of smell and taste with no other symptoms."
]

for text in texts:
    tokens, attention_matrix = get_attention_scores(text)
    print(f"Text: {text}\nTokens: {tokens}\n")
    plot_attention_heatmap(tokens, attention_matrix)
```
